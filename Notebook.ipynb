```python
# Pandas es una librería muy poderosa para el manejo y análisis de datos en Python.
import pandas as pd
# Matplotlib es una librería de visualización en Python.
import matplotlib.pyplot as plt
# Seaborn es una librería de visualización basada en matplotlib
import seaborn as sns
```

```python
# Está cargando un archivo CSV llamado "VNL2023.csv" en un DataFrame de pandas. 
# Esto significa que estás leyendo los datos de ese archivo y almacenándolos en la variable df, 
# que es una estructura de pandas utilizada para manejar datos tabulares
df = pd.read_csv("VNL2023.csv")
```

```python
# Ver las primeras filas del DataFrame
df.head()
```

```python
# Número de filas y columnas
print(df.shape)
```

```python
# Ver estadísticas descriptivas
df.describe()
```

```python
# Nº valores nulos por columna
df.isna().sum()
```

```python
# Nº filas duplicadas
df.duplicated().sum()
```

```python
# Lista de las columnas que contienen valores numéricos (enteros y flotantes)
numeric_cols = df.select_dtypes(include =["int","float"]).columns
# Matriz de correlación de las columnas seleccionada
# Matriz de correlación es una tabla que muestra el coeficiente de correlación de Pearson 
# entre cada par de variables numéricas. Los valores estarán entre -1 y 1
corr_matrix = df[numeric_cols].corr()
# Imprimir
print(corr_matrix)
```

```python
# Mapa de calor de Seaborn con matriz de correlación apartado anterior
# Sí a mostrar valores nº dentro de celda, paleta de color y grosor líneas
sns.heatmap(corr_matrix, annot= True, cmap = "coolwarm", linewidths= 5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Nº veces cada valor único aparece en columna
position_counts = df["Position"].value_counts()
position_counts
```

```python
# Tipo de dato del objeto
type(position_counts)
```

```python
# Filas del DataFrame
position_counts.index
```

```python
# Crear gráfico circular pie chart
# Asignamos etiquetas al gráfico iguales a índices de position_counts
# 1 decimal
# Rotamos para primer segmento empiece en 90º
plt.pie(position_counts, labels = position_counts.index, autopct = "%1.1f%%", startangle=90)
plt.title("Distribution of Positions")
plt.show()
```

```python
# Sin porcentajes
plt.pie(position_counts, labels = position_counts.index, startangle=90)
plt.title("Distribution of Positions")
plt.show()
```

```python
# Agrupamos por países y devolvemos la media de cada uno en "Attack" descendentemente
avg_attack_by_country = df.groupby("Country")["Attack"].mean()
avg_attack_by_country.sort_values(ascending=False)
```

```python
# Gráfico de barras descendente por orden de valores de ataque medio por país
avg_attack_by_country.sort_values(ascending=False).plot(kind="bar")
plt.title("Average Attack by Country")
plt.xlabel("Country")
plt.ylabel("Average Attack")
plt.show()
```

```python
# Media de valores columna "Serve" agrupados por los diferentes valores de "Age"
# Mostrado de mayor a menor
avg_serve_by_age = df.groupby("Age")["Serve"].mean()
avg_serve_by_age.sort_values(ascending=False)
```

```python
# DataFrame con las 20 combinaciones de "Country" y "Position" que tienen los valores más altos en "Attack"
df.groupby(["Country", "Position"])["Attack"].max().reset_index().sort_values(ascending=False, by = "Attack").head(20)
```

```python
# Ahora mostramos las 20 últimas filas con tail
df.groupby(["Country", "Position"])["Attack"].max().reset_index().sort_values(ascending=False, by = "Attack").tail(20)
```

```python
# Calculamos la suma de "Dig" para cada país
df.groupby("Country")["Dig"].sum()
```

```python
# Cambiamos nombre a la columna
df.rename(columns = {"Recieve ": "Receive"}, inplace = True)
```

```python
# Gráfico de dispersión de Matplotlib
plt.scatter(df["Block"], df["Receive"])
plt.title("Scatter Plot : Block vs Receive")
plt.xlabel("Block")
plt.ylabel("Receive")
plt.show()
```

```python
# Diagrama de caja de Seaborn
sns.boxplot(x = df["Serve"])
plt.title("Box Plot: Distribution of Serve Values")
plt.xlabel("Serve")
plt.show()
```

```python
# Historigrama con Matplotlib
plt.hist(df["Age"], bins = 20, color = "skyblue", edgecolor = "black")
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Media de ataque por posición
# Ordenado ascendentemente
avg_attack_by_position = df.groupby("Position")["Attack"].mean()
avg_attack_by_position.sort_values(ascending = True).plot(kind = "bar", color = "green")
plt.title("Average Attack by Position")
plt.xlabel("Position")
plt.ylabel("Average Attack")
plt.show()
```

```python
# Agrupamos Media de "Serve" por edad
serve_trend_by_age = df.groupby("Age")["Serve"].mean()
# Gráfico de línea
# Marcadores círculos, línea sólida, color azul
serve_trend_by_age.plot(kind = "line", marker = "o", linestyle = "-", color = "blue")
plt.title("Serve Trend Over Age Groups")
plt.xlabel("Age")
plt.ylabel("Average Serve")
plt.show()
```

```python
# Agrupamos países y mostramos suma de su "Attack" y "Block"
total_attack_by_country = df.groupby("Country")[["Attack","Block"]].sum()
```

```python
# Lo mostramos
total_attack_by_country
```

```python
# Gráfico de barras apiladas con Matplotlib
# Mostramos valores de "Attack" y "Block" por país
# Colormap "viridis"
# stacked = True para apilar barras
total_attack_by_country.sort_values(ascending=False,by="Attack").plot(kind = "bar", stacked = True, colormap="viridis")
plt.title("Total Attack and Block by Country")
plt.xlabel("Country")
plt.ylabel("Total Values")
plt.show()
```
