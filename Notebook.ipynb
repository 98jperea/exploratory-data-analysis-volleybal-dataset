```python
# Pandas es una librería muy poderosa para el manejo y análisis de datos en Python.
import pandas as pd
# Matplotlib es una librería de visualización en Python.
import matplotlib.pyplot as plt
# Seaborn es una librería de visualización basada en matplotlib
import seaborn as sns
```

```python
# Está cargando un archivo CSV llamado "VNL2023.csv" en un DataFrame de pandas. 
# Esto significa que estás leyendo los datos de ese archivo y almacenándolos en la variable df, 
# que es una estructura de pandas utilizada para manejar datos tabulares
df = pd.read_csv("VNL2023.csv")
```

```python
# Ver las primeras filas del DataFrame
df.head()
```

```python
# Número de filas y columnas
print(df.shape)
```

```python
# Ver estadísticas descriptivas
df.describe()
```

```python
# Nº valores nulos por columna
df.isna().sum()
```

```python
# Nº filas duplicadas
df.duplicated().sum()
```

```python
# Lista de las columnas que contienen valores numéricos (enteros y flotantes)
numeric_cols = df.select_dtypes(include =["int","float"]).columns
# Matriz de correlación de las columnas seleccionada
# Matriz de correlación es una tabla que muestra el coeficiente de correlación de Pearson 
# entre cada par de variables numéricas. Los valores estarán entre -1 y 1
corr_matrix = df[numeric_cols].corr()
# Imprimir
print(corr_matrix)
```

```python
# Mapa de calor de Seaborn con matriz de correlación apartado anterior
# Sí a mostrar valores nº dentro de celda, paleta de color y grosor líneas
sns.heatmap(corr_matrix, annot= True, cmap = "coolwarm", linewidths= 5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Figura 20 ancho, 12 alto
plt.figure(figsize = (20, 12))
# Mapa de calor de Seaborn con matriz de correlación apartado anterior
# Sí a mostrar valores nº dentro de celda, paleta de color y grosor líneas
sns.heatmap(corr_matrix, annot= True, cmap = "coolwarm", linewidths= 5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# Nº veces cada valor único aparece en columna
position_counts = df["Position"].value_counts()
position_counts
```

```python
# Tipo de dato del objeto
type(position_counts)
```

```python
# Filas del DataFrame
position_counts.index
```

```python
# Crear gráfico circular pie chart
# Asignamos etiquetas al gráfico iguales a índices de position_counts
# 1 decimal
# Rotamos para primer segmento empiece en 90º
plt.pie(position_counts, labels = position_counts.index, autopct = "%1.1f%%", startangle=90)
plt.title("Distribution of Positions")
plt.show()
```

```python
# Sin porcentajes
plt.pie(position_counts, labels = position_counts.index, startangle=90)
plt.title("Distribution of Positions")
plt.show()
```

```python
# Agrupamos por países y devolvemos la media de cada uno en "Attack" descendentemente
avg_attack_by_country = df.groupby("Country")["Attack"].mean()
avg_attack_by_country.sort_values(ascending=False)
```
